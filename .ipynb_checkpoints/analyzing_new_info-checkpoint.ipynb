```python
df.info()
```

```python
# We're going format the job_posted_date into a datetime 
df['job_posted_date'] = pd.to_datetime(df.job_posted_date)
```

```python
# Let's extract the month from the job_posted_date column to create a special column of months  
df['job_posted_month'] = df.job_posted_date.dt.month
```

```python
df.info()
```

```python
# df_multiple[(df_multiple.job_title_short == 'Data Analyst') & (df_multiple.salary_year_avg.notna())]
df.dropna(subset=['salary_year_avg'], inplace=True)
```

```python
df
```

```python
# Let's get rid of the salary_hour_avg column using the Drop method
df.drop(labels='salary_hour_avg', axis=1, inplace=True)
```

```python
# Let's just combine them
job_title_grouped_comb = df.groupby('job_title_short')['salary_year_avg'].agg(['min', 'median', 'max'])
# job_title_grouped_comb = df.groupby(['job_title_short', 'job_location'])['salary_year_avg'].agg(['min', 'median', 'max'])
job_title_grouped_comb
```

```python
job_title_grouped_comb.plot(kind='line')
plt.ylabel('Average Salary Per Year')
plt.xlabel('')
plt.xticks(rotation=45, ha='right')
plt.show()
```
